## Partial Equilibrium Dynamics in Bewley-Aiyagari Models

```julia
include("incomplete_marketsGE.jl")
```

```julia
a = geomspace(0., 200., 500)
p = Params()
y, pr, Π = markov_rouwenhorst(p.ρ, p.σ, N=7);
up(c) = 1 ./ c
up_inv(c) = 1 ./ c

r = 0.01/4

c, a₊, D, C, A = ss(a, y, r, Π, up, up_inv, p; verbose = false);
```

### 1% shock to interest rate at $t=0$

```julia
rs =  r .+ 0.01 .* (collect(1:10) .== 1)
Cs, As = td_PE(a, y, r, Π, up, up_inv, p, rs);
```

```julia
plot((Cs[:] .- C)/C, lw = 2, legend = false, title = "Consumption Response" )
```

```julia
plot((As[:] .- A)/A, lw = 2, legend = false, title = "Asset Response" )
```

### 1% shock to r at date $t=5$

```julia
rs =  r .+ 0.01 .* (collect(1:20) .== 5)
Cs, As = td_PE(a, y, r, Π, up, up_inv, p, rs);
```

```julia
plot((Cs[:] .- C)/C, lw = 2, legend = false, title = "Consumption Response" )
```

### 1% proportional shock to income
In this case response is driven by the income-weighted MPC

```julia
ys = y' .+ y' .* (0.01 *  (collect(1:20) .== 1))
Cs, As = td_PE(a, y, r, Π, up, up_inv, p, ys = ys);
```

```julia
plot((Cs[:] .- C)/C, lw = 2, legend = false, title = "Consumption Response" )
```

```julia
ys = y' .+ y' .* (0.01 *  (collect(1:30) .== 10))
Cs, As = td_PE(a, y, r, Π, up, up_inv, p, ys = ys);
```

```julia
plot((Cs[:] .- C)/C, lw = 2, legend = false, title = "Consumption Response" )
```
